# 3.Data Visualization

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

## The following histograms show the distribution of key weather parameters - that is Temperature and Humidity

```python
sns.histplot(df['Temp_C'],kde=True)
plt.title('Temperature Distribution')
plt.show()
```

```python
sns.histplot(df['Rel Hum_%'], kde=True)
plt.title("Humidity Distribution")
plt.show()
```

## Plot time series graphs to visualize Dew Point Temp_C and Temp_C over time

```python

y = df["Dew Point Temp_C"]
x = df["Date/Time"]
plt.scatter(x,y)
plt.title("Scatter Plot of Dew Point Temp_C vs Date/Time")
plt.xlabel("Date/Time")
plt.ylabel("Dew Point Temp_C")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df=pd.read_csv(r'C:\Users\LAMECK OMENDA\Downloads\Weather Data.csv')

plt.figure(figsize=(10,6))
plt.plot(df['Date/Time'],df['Temp_C'])
plt.title('Temp_C over Time')
plt.xlabel('Date/Time')
plt.ylabel('Temp_C')
plt.show()
```

Both Dew Point Temp_C and Temp_C have the same seasonal variation. Their values started increasing then reaching a particular point they 
started decreasing

## Correlation matrices and heatmaps identifying different weather parameters.

```python
df_numeric= df.select_dtypes(include=[float,int])
corr_matrix = df_numeric.corr()
corr_matrix
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot = True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

# 4. Weather patterns and Trends

```python
df['Date/Time'] = df['Date/Time']
monthly_avg_temp = df.groupby('Date/Time')['Temp_C'].mean()

plt.plot(monthly_avg_temp)
plt.title('Average Monthly Temperature')
plt.xlabel('Date/Time')
plt.ylabel('Temp_C')
plt.show()
```
